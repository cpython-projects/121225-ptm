### 1. Слияние интервалов (Meeting Scheduler)

На входе список кортежей — интервалы времени (например, бронирование переговорки). Нужно объединить все пересекающиеся интервалы.

* **Вход:** `[(1, 3), (2, 6), (8, 10), (15, 18)]`
* **Выход:** `[(1, 6), (8, 10), (15, 18)]`


### 2. "Дырки" в последовательности (Missing ID)

В базе данных есть автоинкрементные ID, но некоторые записи были удалены. Найдите все пропущенные ID в заданном диапазоне.

* **Вход:** `[1, 2, 4, 6, 7]`, диапазон от 1 до 7
* **Выход:** `[3, 5]`



### 3. Скользящее среднее (Monitoring Metrics)

Для системы мониторинга нужно вычислить скользящее среднее с окном  для потока метрик.

* **Вход:** `[10, 20, 30, 40, 50]`, окно `K=3`
* **Выход:** `[20.0, 30.0, 40.0]` (где 20.0 = (10+20+30)/3)


### 4. Удаление дубликатов In-place

Представьте, что список очень большой и нельзя создавать его копию. Удалите дубликаты из **отсортированного** списка, сдвинув уникальные элементы в начало. Верните длину «чистой» части.

* **Вход:** `[1, 1, 2, 2, 2, 3]`
* **Выход:** `3` (список модифицирован в `[1, 2, 3, _, _, _]`)

### 5. Круговой сдвиг (Load Balancer Rotation)

Реализуйте циклическое смещение списка вправо на k шагов. Так часто работают простые алгоритмы распределения задач (Round Robin).

* **Вход:** `[1, 2, 3, 4, 5]`, `K=2`
* **Выход:** `[4, 5, 1, 2, 3]`


### 6. Пересечение множеств (Common Followers)

Найдите общие элементы двух списков без использования типа данных `set`. Попробуйте оптимизировать решение, если списки уже отсортированы (метод двух указателей).

* **Вход:** `[1, 2, 4, 5]` и `[3, 4, 5, 6]`
* **Выход:** `[4, 5]`


### 7. Самая длинная последовательность единиц

Дан массив из 0 и 1. Вы можете заменить один 0 на 1. Какую максимальную длину последовательности из единиц можно получить?

* **Вход:** `[1, 0, 1, 1, 0, 1]`
* **Выход:** `4` (заменив 0 на индексе 1)

### 8. Chunking (Постраничная выборка)

Разбейте один большой список на подсписки фиксированного размера . Это база для реализации пагинации или пакетной вставки в БД (Batch Insert).

* **Вход:** `[1, 2, 3, 4, 5, 6, 7]`, размер `3`
* **Выход:** `[[1, 2, 3], [4, 5, 6], [7]]`


### 9. "Упаковка" дубликатов (Group Consecutive)

Сгруппируйте идущие подряд одинаковые элементы в кортежи вида `(значение, количество)`. Похоже на RLE, но для любых объектов в списке.

* **Вход:** `[1, 1, 1, 0, 0, 2, 1, 1]`
* **Выход:** `[(1, 3), (0, 2), (2, 1), (1, 2)]`

### 10. Префиксные суммы (Balance Checker)

Создайте новый список, где каждый элемент — это сумма всех предыдущих элементов плюс текущий. Помогает мгновенно вычислять сумму на любом отрезке (например, сумма транзакций за период).

* **Вход:** `[1, 2, 3, 4]`
* **Выход:** `[1, 3, 6, 10]`
