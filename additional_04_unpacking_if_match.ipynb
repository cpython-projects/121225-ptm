### Задача 1: Анализатор команд робота

Робот принимает команды в виде списков. Нужно распознать действие и параметры.

* **Вход:** Список, например `["move", 10]`, `["rotate", 90]`, `["stop"]`.
* **Условие:**
1. Если `["move", dist]`, где `dist` — число: вывести "Движение вперед на {dist} метров".
2. Если `["rotate", angle]`: вывести "Поворот на {angle} градусов".
3. Если `["stop"]`: вывести "Робот остановлен".


* **Результат:** Текстовое описание действия или "Ошибка" для всех остальных форматов.


### Задача 2: Обработка HTTP-статусов

Система получает кортеж из кода ответа и описания.

* **Вход:** Кортеж вида `(200, "OK")`, `(404, "Not Found")`, `(500, "Server Error")`.
* **Условие:**
1. Если код начинается на `2` (например, 200, 201): вывести "Успешно: {описание}".
2. Если код `404`: вывести "Страница не найдена".
3. Если код — любое другое число: вывести "Статус: {код}".


* **Результат:** Строка с классификацией ответа.


### Задача 3: Валидация геометрических фигур

Данные приходят в виде списка, где первый элемент — название фигуры, а остальные — размеры.

* **Вход:** `["circle", 5]`, `["rect", 10, 20]`, `["triangle", 3, 4, 5]`.
* **Условие:**
1. Если круг (`circle`) и есть один радиус: вывести "Круг с радиусом {r}".
2. Если прямоугольник (`rect`) и есть два значения: вывести "Прямоугольник {w}x{h}".
3. Если три значения (неважно, какое название фигуры): вывести "Фигура с 3 сторонами".


* **Результат:** Описание фигуры.


### Задача 4: Умный сумматор (Типы данных)

Функция принимает список из двух элементов и пытается их сложить, только если они совместимы.

* **Вход:** `[10, 20]`, `["Hello ", "World"]`, `[5, "test"]`.
* **Условие:**
1. Если оба элемента — `int`: вывести их сумму.
2. Если оба элемента — `str`: вывести их конкатенацию.
3. Во всех остальных случаях: вывести "Несовместимые типы".


* **Результат:** Результат сложения или сообщение об ошибке.


### Задача 5: Парсинг путей файловой системы

Дан список строк, представляющий путь. Нужно определить «глубину» и тип доступа.

* **Вход:** `["root", "users", "admin"]`, `["root", "etc"]`, `["root"]`.
* **Условие:**
1. Если путь начинается с `"root"` и в нем ровно 3 элемента: вывести "Доступ к пользовательской директории".
2. Если путь состоит только из одного элемента `"root"`: вывести "Корень системы".
3. Если путь начинается с `"root"`, а дальше идет любое количество элементов: вывести "Путь внутри root: {остальные_элементы}".


* **Результат:** Категория пути.



### Задача 6: Система уведомлений (Wildcards)

Уведомление приходит как список: `[уровень_важности, сообщение, *доп_инфо]`.

* **Вход:** `["INFO", "Update"]`, `["ERROR", "Crash", "log.txt", "12:00"]`.
* **Условие:**
1. Если первый элемент `"ERROR"`, собрать все элементы после сообщения в список `details`: вывести "Ошибка: {сообщение}, детали: {details}".
2. Если первый элемент `"INFO"` и элементов всего два: вывести "Инфо: {сообщение}".


* **Результат:** Отформатированное сообщение.


### Задача 7: Обработка банковских транзакций

Транзакция — это список `[тип, сумма, валюта]`. Но иногда валюта не указана.

* **Вход:** `["deposit", 100, "USD"]`, `["withdraw", 50]`.
* **Условие:**
1. Если тип `"deposit"` и есть сумма и валюта: вывести "Пополнение на {сумма} {валюта}".
2. Если указан тип и только сумма (валюта отсутствует): считать валютой по умолчанию "USD" и вывести "Операция: {тип}, сумма: {сумма} USD".


* **Результат:** Описание транзакции с подстановкой валюты.


### Задача 8: Светофор (Оператор ИЛИ)

На вход подается текущий цвет и время до переключения.

* **Вход:** `("red", 10)`, `("green", 5)`, `("yellow", 2)`.
* **Условие:**
1. Если цвет `"red"` или `"yellow"`: вывести "Стойте".
2. Если цвет `"green"` и времени больше 3 секунд: вывести "Идите".
3. Если цвет `"green"` и времени 3 и меньше: вывести "Ускорьтесь или приготовьтесь стоять".


* **Результат:** Команда для пешехода.


### Задача 9: Распаковка вложенных списков

Дан список, в котором может быть еще один вложенный список.

* **Вход:** `[1, [2, 3]]`, `[1, 2]`.
* **Условие:**
1. Если структура `[a, [b, c]]`: вывести сумму всех трех чисел.
2. Если структура `[a, b]`: вывести произведение этих двух чисел.


* **Результат:** Число (результат вычисления).

### Задача 10: Логика "Админ-Панели"

Дан список `[имя_пользователя, роль, статус_бана]`.

* **Вход:** `["Alex", "admin", False]`, `["Bot", "user", True]`.
* **Условие:**
1. Если роль `"admin"` независимо от бана: вывести "Полный доступ".
2. Если роль `"user"` и статус бана `False`: вывести "Обычный доступ".
3. Если статус бана `True`: вывести "Доступ заблокирован".


* **Результат:** Вердикт по доступу.
